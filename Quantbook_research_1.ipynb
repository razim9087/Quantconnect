```python
# QuantBook Analysis Tool
# For more information see [https://www.quantconnect.com/docs/v2/our-platform/research/getting-started]
qb = QuantBook()
spy = qb.add_equity("SPY")
aapl = qb.add_equity("AAPL")
goog = qb.add_equity("GOOG")
```

Price History

```python
history = qb.history(qb.securities.keys(), 365, Resolution.DAILY)
```

```python
history
```

```python
history.loc[spy.symbol]
```

Fundamental Data

```python
ebit_history=qb.GetFundamental(qb.Securities.Keys,
                               "FinancialStatements.IncomeStatement.EBIT.TwelveMonths",
                               datetime(2024,1,1),datetime(2025,12,31))
ebit_history
```

```python
ebit_history.columns
```

```python
ebit_history[ebit_history.columns[0]]
```

```python
ebit_history[ebit_history.columns[0]].resample("Q").mean()
```

Technical Indicators

```python
# Indicator Analysis
bbdf = qb.indicator(BollingerBands(30, 2), spy.symbol, 360, Resolution.DAILY)
bbdf.drop('standarddeviation', axis=1).plot()
```
